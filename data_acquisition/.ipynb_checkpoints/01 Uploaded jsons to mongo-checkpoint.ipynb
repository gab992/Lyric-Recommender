```python
from pymongo import MongoClient
from pprint import pprint
import os
import json
```

```python
# Connecting to mongoDB

config = {
  'host': '',
  'username': '',
  'password': '',
  'authSource': ''
}

client = MongoClient(**config)

db = client.songlyrics
```

```python
# Checking for collections in songlyrics

db.collection_names()
```

```python
# Creating collection for songs

songs = db['songs']
```

```python
# Addings non-translation lyrics to database and saving a list of lyrics that are translations

translations = []
for file in os.listdir('.'):
    if file.startswith('Genius'):
        translations.append(file)
    elif file.endswith('.json'):
        with open(file) as f:
            file_data = json.load(f)
            db.songs.insert_one(file_data)
```

```python
# Checking that data upload worked
cursor = db.songs.find().limit(4)
list(cursor)
```
